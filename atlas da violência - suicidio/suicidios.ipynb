# Atlas da violência - suicídios

fontes: [Instituto de Pesquisa Econômica Aplicada](https://www.ipea.gov.br/atlasviolencia/filtros-series)

```python
import pandas as pd
import matplotlib.pyplot as plt

suicidioMUF = pd.read_csv('./data/suicidioMUF.csv', sep=";")
suicidioMRE = pd.read_csv("./data/suicidioMRE.csv", sep=";")
suicidioHUF = pd.read_csv('./data/suicidioHUF.csv', sep=";")
suicidioHRE = pd.read_csv("./data/suicidioHRE.csv", sep=";")
regioesNome =  {"SE":"Sudeste", "CO":"Centro-Oeste", "N":"Norte", "S":"Sul", "NE":"Nordeste"}
```

## Suicídio de mulheres

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), dpi=90)
estados = suicidioMUF["nome"].unique()
regioes = suicidioMRE["nome"].unique()

for estado in estados:
  data = suicidioMUF.loc[suicidioMUF["nome"] == estado]
  x = data["período"].unique()[0:9]
  y = data["valor"].unique()[0:9]
  ax2.plot(x, y, label=estado)
  ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))

for regiao in regioes:
  data1 = suicidioMRE.loc[suicidioMRE["nome"] == regiao]
  z = data1["período"].unique()[0:9]
  k = data1["valor"].unique()[0:9]
  ax1.plot(z, k, label=regioesNome[regiao])
  ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax1.set_ylabel("Suicídios", size=20)
fig.tight_layout(pad=5.0)
plt.show()
```

## Suicídio de homens

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10), dpi=90)
estados = suicidioMUF["nome"].unique()
regioes = suicidioMRE["nome"].unique()

for estado in estados:
  data = suicidioHUF.loc[suicidioHUF["nome"] == estado]
  x = data["período"].unique()[0:9]
  y = data["valor"].unique()[0:9]
  ax2.plot(x, y, label=estado)
  ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))

for regiao in regioes:
  data1 = suicidioHRE.loc[suicidioHRE["nome"] == regiao]
  z = data1["período"].unique()[0:9]
  k = data1["valor"].unique()[0:9]
  ax1.plot(z, k, label=regioesNome[regiao])
  ax1.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax1.set_ylabel("Suicídios", size=20)
fig.tight_layout(pad=5.0)
plt.show()
```
